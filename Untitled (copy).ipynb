```python
# Define max_distance (eps parameter in DBSCAN())
db = DBSCAN(eps=1.7, min_samples=25).fit(df)

# Extract a mask of core cluster members
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

# Extract labels (-1 is used for outliers)
labels = db.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_cluster = n_clusters + 1
unique_labels = set(labels)
n_noise_ = list(labels).count(-1)
```

```python
print('Estimated number of clusters: %d' % n_clusters)
print('Total Data : ', len(df))
print('Estimated number of noise points: %d' % n_noise_)
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(df, labels))
```

```python
%matplotlib inline

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == 1:
        # block used noise.
        col = [0, 0, 0, 1]
        
    class_member_mask = (labels == k)

    xy = df[class_member_mask & core_samples_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=10)
    print('\n', xy,'\nMember of Cluster : ', len(xy))

    xy = df[class_member_mask & ~core_samples_mask]
    plt.plot(xy['longitude'], xy['latitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=3)
    
min_x = np.min(df['longitude'])
max_x = np.max(df['longitude'])
min_y = np.min(df['latitude'])
max_y = np.max(df['latitude'])
    
    
# Plot up the results!
# fig = plt.figure()
plt.xlim(min_x, max_x)
plt.ylim(min_y, max_y)
plt.ylabel('Posisi longitude')
plt.xlabel('Posisi latitude')
plt.title('DBSCAN: %d clusters found' % n_clusters, fontsize = 20)
# fig.tight_layout()
plt.show()
```
